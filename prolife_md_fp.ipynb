```python
# convert mol2 files for the proteins to pdb format using open bable
os.chdir('C:/Users/sbu-e12/Downloads/CmDock_DUD-E_DB')

os.getcwd()
ligand_list = os.listdir()


for lig in ligand_list:
    os.chdir(f'C:/Users/sbu-e12/Downloads/CmDock_DUD-E_DB/{lig}')

    ! obabel rec_{lig}.mol2 -O rec_{lig}.pdb
    os.chdir('C:/Users/sbu-e12/Downloads/CmDock_DUD-E_DB')
```

```python
# generate fingerptints

for lig in ligand_list:
    print(lig)
    os.chdir(f'C:/Users/sbu-e12/Downloads/CmDock_DUD-E_DB/{lig}')
    try:
        # read protein
        protein_file = f'rec_{lig}.pdb'
        rdkit_prot = Chem.MolFromPDBFile(protein_file, removeHs=False)
        rdkit_protein_mol = plf.Molecule(rdkit_prot)

        # read SDF
        ligand_file = 'results.sdf'
        ligand_mol = plf.sdf_supplier(ligand_file)[0]

        # use default interactions
        fp = plf.Fingerprint()
        # run on your poses
        fp.run_from_iterable([ligand_mol], rdkit_protein_mol)

        # save pickle for fingerprint
        fp.to_pickle("fingerprint.pkl")
        fp = plf.Fingerprint.from_pickle("fingerprint.pkl")

        df = fp.to_dataframe()
        df.to_csv('fp.csv')
    except:
        'ArgumentError'
```
